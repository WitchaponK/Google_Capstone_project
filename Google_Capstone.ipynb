### Logistic Regression : satisfaction & stayed/left

```python
X = df2[selected_features]
y= df2['left']
```

```python
X_temp, X_test, y_temp, y_test = train_test_split(X,y,test_size = 0.20, stratify = y, random_state = random_state )

X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size = 0.25, stratify = y_temp, random_state = random_state )
```

```python
for x in [X_temp, X_train, X_test, X_val]:
    print(len(x))
```

```python
lr_clf =  LogisticRegression(random_state = random_state, max_iter = 500).fit(X_train,y_train)
ylr_pred = lr_clf.predict(X_val)
```

```python
print(lr_clf.coef_)
print(lr_clf.intercept_)
```

```python
import sklearn.metrics as metrics
```

```python
lr_cm = confusion_matrix(y_val, ylr_pred)
lr_disp = ConfusionMatrixDisplay(confusion_matrix = lr_cm, display_labels = ['stayed','left'])
lr_disp.plot(values_format=' ')
plt.show()
```

```python
target_names = ['Predicted would not leave', 'Predicted would leave']
print(classification_report(y_val, ylr_pred, target_names=target_names))
```

```python
score_table = pd.DataFrame(columns = ['model_name','accuracy','precision','recall','f1_score'])
score_table
```

```python
logistic_regression_val_score = pd.DataFrame({'model_name':'logistic_regression_val_score',
                            'accuracy':[accuracy_score(y_val, ylr_pred)],
                            'precision':[precision_score(y_val, ylr_pred)],
                            'recall':[recall_score(y_val, ylr_pred)],
                            'f1_score':[f1_score(y_val, ylr_pred)]}
                          )
score_table = pd.concat([score_table,logistic_regression_val_score], ignore_index = True, axis = 0)
score_table
```

Logistic regression has good accuray(83.0%) , mediocore precision(47.2%) and very low recall(18.8%)\
but remember, we have imbalance data set which will impact the accuracy significantly, since the model\
would naturally able to predict the people who stay (whether they truly stay or not, but since most of them\

### Decision Tree Modeling
Next we use decision tree, and we want to identify who we predicted 'left' and actually left (true positive) , and minimize those who actually left but predicted 'stay' (false negative)\
so, we will use 'recall' as our model measure

```python
#index the validation data set as 0, and -1 for training data set

split_index = [0 if x in X_val.index else -1 for x in X_temp.index]


from sklearn.model_selection import PredefinedSplit

custom_split = PredefinedSplit(split_index)
#custom_split will be index for validation data set when using GridsearchCV
```

```python
%%time

dt_val = DecisionTreeClassifier(random_state = random_state)
dt_val_para = {'max_depth':[4,5,6,7,8,9,10,11,12,15,20,30,40,50],
             'min_samples_leaf': [2, 5, 10, 20, 50]}
scoring = ['accuracy', 'precision', 'recall', 'f1']



dt_val_cv  = GridSearchCV(dt_val, dt_val_para, scoring = scoring, cv=custom_split, refit="recall")



dt_val_cv.fit(X_temp, y_temp)
```

```python
print(dt_val_cv.best_estimator_)
print(dt_val_cv.best_score_)
print(dt_val_cv.best_params_)
```

```python
dt_val_pred = dt_val_cv.predict(X_val)
```

```python
decision_tree_val_score = pd.DataFrame({'model_name':'decision_tree_val_score',
                            'accuracy':[accuracy_score(y_val, dt_val_pred)],
                            'precision':[precision_score(y_val, dt_val_pred)],
                            'recall':[recall_score(y_val, dt_val_pred)],
                            'f1_score':[f1_score(y_val, dt_val_pred)]}
                          )
score_table = pd.concat([score_table,decision_tree_val_score], ignore_index = True, axis = 0)
score_table
```

Decision tree result is significantly better than logistic regression with very high accuracy at 97.6%\
 high precision, f1 and recall at 92.7%, 92.3% and 91.9% respectively
